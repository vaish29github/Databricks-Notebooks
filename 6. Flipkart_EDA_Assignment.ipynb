```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Flipkart Data Engineering").getOrCreate()
file_path = '/FileStore/tables/Flipkart.csv'
flipkart_df = spark.read.csv(file_path, header=True, inferSchema=True)
```

```python
flipkart_df.printSchema()
flipkart_df.count()
```

```python
flipkart_df.show(5)
```

```python
# Display all category name
flipkart_df.select("maincateg").distinct().display(truncate=False)
```

```python
# Filter products with rating > 4.5 and more than 1000 reviews
filtered_products = flipkart_df.filter((flipkart_df.Rating > 4) & (flipkart_df.noreviews1 > 100))
filtered_products.display(truncate=False)
```

```python
# Display Products in 'Men' category that are fulfilled
men_fulfilled_products = flipkart_df.filter((flipkart_df.maincateg == 'Men') & (flipkart_df.fulfilled1 == 1))
men_fulfilled_products.display(truncate=False)
```

```python
# Dsiplay number of products per category
category_counts = flipkart_df.groupBy("maincateg").count()
category_counts.display(truncate=False)
```

```python
# Display Average rating per category
from pyspark.sql.functions import avg

avg_rating_per_category = flipkart_df.groupBy("maincateg").agg(avg("Rating").alias("avg_rating"))
avg_rating_per_category.display(truncate=False)
```

```python
# Dsiplay Category with highest average number of reviews
from pyspark.sql.functions import avg

highest_avg_reviews = (
    flipkart_df.groupBy("maincateg")
    .agg(avg("noreviews1").alias("avg_reviews"))
    .orderBy("avg_reviews", ascending=False)
    .limit(1)
)

highest_avg_reviews.display(truncate=False)
```

```python
# Top 5 products with highest price. display product name and price
top_5_expensive = (
    flipkart_df.select("title", "actprice1")
    .orderBy("actprice1", ascending=False)
    .limit(5)
)

top_5_expensive.display(truncate=False)
```

```python
# Display Min, max, and avg price per category


from pyspark.sql.functions import min, max, avg

price_stats = (
    flipkart_df.groupBy("maincateg")
    .agg(
        min("actprice1").alias("min_price"),
        max("actprice1").alias("max_price"),
        avg("actprice1").alias("avg_price")
    )
)

price_stats.display(truncate=False)
```

```python
# Display number of nulls in each column
from pyspark.sql.functions import col, sum

null_counts = flipkart_df.select([
    sum(col(c).isNull().cast("int")).alias(c + "_nulls") for c in flipkart_df.columns
])

null_counts.display()
```

```python
# calculate and display the category name, number of fulfilled, and unfulfilled products
from pyspark.sql.functions import when, count

flipkart_df.groupBy("maincateg") \
  .agg(
    count(when(flipkart_df["fulfilled1"] == 1, 1)).alias("fulfilled_count"),
    count(when(flipkart_df["fulfilled1"] == 0, 1)).alias("unfulfilled_count")
  ) \
  .display(truncate=False)
```

```python
# Display Count of products per category
category_product_count = flipkart_df.groupBy("maincateg").count()
category_product_count.display(truncate=False)
```

```python
# Display Average rating per category
# Display Category with highest average number of reviews

from pyspark.sql.functions import avg

avg_rating_per_category = (
    flipkart_df.groupBy("maincateg")
    .agg(avg("Rating").alias("avg_rating"))
)

avg_rating_per_category.display(truncate=False)
```

```python
# Display Category with highest average number of reviews

from pyspark.sql.functions import avg

highest_avg_reviews = (
    flipkart_df.groupBy("maincateg")
    .agg(avg("noreviews1").alias("avg_reviews"))
    .orderBy("avg_reviews", ascending=False)
    .limit(1)
)

highest_avg_reviews.display(truncate=False)
```

```python
# Display Bar chart of product count per category

# Use Databricks UI to visualize this as a bar chart
category_counts = flipkart_df.groupBy("maincateg").count()
category_counts.display()
```

```python
# Bar chart of average rating per category

from pyspark.sql.functions import avg

avg_rating_df = (
    flipkart_df.groupBy("maincateg")
    .agg(avg("Rating").alias("avg_rating"))
)

avg_rating_df.display()
```

```python
# Display Bar chart of total number of reviews per category
from pyspark.sql.functions import sum

total_reviews_per_category = (
    flipkart_df.groupBy("maincateg")
    .agg(sum("noreviews1").alias("total_reviews"))
)

total_reviews_per_category.display()
```

```python
# Display product name and 5 star rating for those products which have highest 5 star rating
from pyspark.sql.functions import col

# Find the maximum 5-star rating
max_5star_rating = flipkart_df.agg({"star_5f": "max"}).collect()[0][0]

# Filter the products with the highest 5-star rating
highest_5star_products = flipkart_df.filter(col("star_5f") == max_5star_rating)

# Display the product name and 5-star rating
highest_5star_products.select("title", "star_5f").display()
```
